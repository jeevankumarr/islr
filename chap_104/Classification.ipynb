## Logistic Regression

```python
x = np.linspace(0,1,500)
#np.log(x/(1-x))
y = np.log(x/(1-x))
traces = []
marker = dict(size=3)
traces.append(go.Scatter(x=x, y=y,mode="lines+markers", marker=marker))
layout = go.Layout(xaxis=dict(title="x"), yaxis=dict(title="log(x/(1-x))"))
fig = go.Figure(data=traces, layout=layout)

py.iplot(fig)
```

### Logistic Model

$$\log \left( \frac{p(X)}{1-p(X)} \right) = \beta_0 + \beta_1X$$

### Estimating the Regression Coefficients

$$l(\beta_0,\beta_1) = \prod_{i:y_{i}=1} p(x_i) \prod_{i^{'}:y_{i^{'}}=0} (1-p(x_{i^{'}}))$$

```python
def_flags = pd.get_dummies(data['default'], prefix='def')
stu_flags = pd.get_dummies(data['student'], prefix='stu')
model_data = pd.merge(data, def_flags, left_index=True, right_index=True)
model_data = pd.merge(model_data, stu_flags, left_index=True, right_index=True)
model_data.head()
```

```python
import statsmodels.api as sm
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

```python
train_cols = ["balance"]
# Index([gre, gpa, prestige_2, prestige_3, prestige_4], dtype=object)
X = model_data[train_cols]
X = sm.add_constant(X)
logit = sm.Logit(model_data['def_Yes'], X)

# fit the model
result = logit.fit()
result.summary()
```

```python
train_cols = ["stu_Yes", "income", "balance"]
# Index([gre, gpa, prestige_2, prestige_3, prestige_4], dtype=object)
X = model_data[train_cols]
X = sm.add_constant(X)
y = model_data['def_Yes']
logit = sm.Logit(y, X)

# fit the model
result = logit.fit()
result.summary()
```

```python
predictions = result.predict(X)
```

```python
th = 0.5
pred_th = [1 if pred >= th  else 0 for pred in predictions]
results = pd.DataFrame({"actual":y, "predictions":pred_th})
results.head()
ct = pd.crosstab(results["actual"], results["predictions"])
display(ct)
opt = []
for th in np.linspace(0.1,0.9,9*4):
    #print("~~~~~~~~~~~{0:0.2f}~~~~~~~~~~~~".format(th))
    pred_th = [1 if pred >= th  else 0 for pred in predictions]
    results = pd.DataFrame({"actual":y, "predictions":pred_th})
    results.head()
    ct = pd.crosstab(results["actual"], results["predictions"])
    #display(ct)

    ct_val = ct.values
    fp = ct_val[0][1]*1.0/(ct_val[0][1]+ct_val[0][0])
    tp = ct_val[1][1]*1.0/(ct_val[1][0]+ct_val[1][1])
    fn = ct_val[1][0]*1.0/(ct_val[1][0]+ct_val[1][1])
    acc = (ct_val[0][0] + ct_val[1][1])*1.0 / np.sum(ct_val)
    err = 1-acc
    #print("fp {0:.2%}, tp {1:.2%}, acc {2:.2%}".format(fp, tp, acc))
    t = {
        "threshold":th,
        "accuracy":acc,
        "error":err,
        "fp_rate":fp,
        "tp_rate":tp,
        "fn_rate":fn
        
    }
    opt.append(t)
opt_df = pd.DataFrame(opt)
opt_df.head()
```

```python
df = opt_df
traces = []

traces.append(go.Scatter(x=df["threshold"], y=df["error"],name="error", mode="lines+markers" ))
traces.append(go.Scatter(x=df["threshold"], y=df["fp_rate"],name="fp_rate", mode="lines+markers" ))
traces.append(go.Scatter(x=df["threshold"], y=df["fn_rate"],name="fn_rate", mode="lines+markers" ))
layout=go.Layout(yaxis=dict(tickformat=".2%", title="rate"), xaxis=dict(title="threshold"))
fig = go.Figure(data=traces, layout=layout)

py.iplot(fig)
```
