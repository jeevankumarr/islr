```python
x = np.random.normal(size=100)
y = np.random.normal(size=100)
```

```python
x
```

```python
traces = []
traces.append(go.Scatter(x=x, y=y, opacity=1, mode="markers"))
py.iplot(traces)
```

```python
traces = []
traces.append(go.Histogram(x=x, opacity=0.50))
traces.append(go.Histogram(x=y, opacity=0.50))
layout = go.Layout(barmode='overlay')
fig = go.Figure(data=traces, layout=layout)

py.iplot(fig)
```
